```python
def highlight_max(s):
    """Highlight maximum value for number of items seen on the row"""
    is_max = s == s.max()
    return ['background-color:yellow' if v else "" for v in is_max]

df.pivot_table(index=['year'], aggfunc='count').style.apply(highlight_max, axis=1)
```

# Cylinders and Displacement


```python
# First make a sorted list of cylinders label
sorted_cylinders_label = df.cylinders_label.value_counts().sort_index().index
sorted_cylinders_label
```

```python
# Scatter plot first: Doesn't help much
ax = df.plot(kind="scatter", x="cylinders", y="displacement")
ax.set_title("Scatter plot of Cylinders vs. Displacement");
```

```python
# Seaborn Stip plot helps. Displacement for 4 cylinders < 6 cylinder 
ax = sns.stripplot(data=df, x='cylinders_label', y='displacement', order=sorted_cylinders_label);
ax.set_title("Seaborn sript plot of Cylinders vs. Displacement with JItter");
```

```python
# Seaborn Stip plot helps, but box plot is better. Black line inside box plot is the median at 50th percentile
# Box 4 cylinder is between 25 and 75th percentile . THe long bars ,mark the upper and lower extremens for all data points.
# The median line is in center, then data points are equally distributed . If the whiskers is long and median is not at center
# , it indicates some Skew . Finally notch can be added around median marker. If notches between 2 cylinders don't overlap, then
# Statistically we can have a test to prove that the two data's don't overlap and were different

ax = sns.boxplot(data=df, x='cylinder_label', y='displacement', order=sorted_cylinders_label, notch=True);
ax.set_title("Seaborn Box plot of Cylinders vs. Displacement with JItter");
```

# Text Analysis

```python
df.head()
```

```python
# Using the str finction, we can access str functions, split on whitespace taking the first item (n=1)
# Value counts summary for the top 10 Brands
ser_car_makes = df.name.str.lower().str.split(n=1, expand=True)[0]
ser_car_makes.value_counts()[:10]
```

```python
# Least frequent 10 Brands
ser_car_makes.value_counts()[-10:]
```

```python
# Mistakes in the data usually come to the bottom counts. Here you can:
#1 Go to source data to fix them
#2 Exclude these incorrect data rows.
# Clean intermediate Python data cleaning script to fix errors and drops mistaken rows from the scraped data.This will give another 
# file ready for analysis.

# Create another pivot table. Pivot on territory and new car_makes and aggregate by size of the group. This counts the number
# of entries that match the pair of territory and car_makes

df['car_makes'] = ser_car_makes
ser_cars_by_territory = df[['territory','car_makes']].pivot_table(index=['territory','car_makes'], aggfunc='size')
df_cars_by_territory = pd.DataFrame(ser_cars_by_territory) # anonymous Series
df_cars_by_territory.columns = ['size']    # Rename anonymous column 0 to named column
df_cars_by_territory = df_cars_by_territory.sort_values(by=['territory','size'], ascending=False)
df_cars_by_territory
```

```python
# Apply lambda filter function to select more common vehicles
mask = df_cars_by_territory.apply(lambda x: x['size'] > 10 , axis=1)
df_cars_by_territory[mask]
```
